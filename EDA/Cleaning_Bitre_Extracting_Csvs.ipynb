# Transforming BITRE Excel Sheets to CSVs

```python
import pandas as pd
import numpy as np
```

## Bitre Data

### International Airline Activity Seat Utilisation

```python
BITRE_REPO = "../raw-data/BITRE_Airline_Activity/"

seat_utilisation_path = 'international_airline_activity_seat_utilisation.xlsx'
seat_utilisation = pd.read_excel(BITRE_REPO+seat_utilisation_path, sheet_name='Data')

# Get header columns
headers = seat_utilisation.iloc[3]
seat_utilisation.columns = headers

seat_utilisation = seat_utilisation.iloc[4:]

# set the index to the date column
seat_utilisation['MONTH'] = pd.to_datetime(seat_utilisation['MONTH'])
seat_utilisation['MONTH'] = seat_utilisation['MONTH'].dt.date
seat_utilisation.set_index('MONTH', inplace=True)

seat_utilisation.columns = ['AIRLINE', 'COUNTRY', 'FLTS IN', 'PAX IN', 'SEATS IN', 'FLTS OUT',
       'PAX OUT', 'SEATS OUT', 'FIN_Y', 'CAL_Y']

seat_utilisation.head()
```

```python
seat_utilisation.to_csv("../raw-csvs/bitre/seat_utilisation.csv")
```

### International Airline Activity 

```python
BITRE_REPO = "../raw-data/BITRE_Airline_Activity/"

airline_activity_path = 'internation_airline_activity_2009_2025.xlsx'
airline_activity = pd.read_excel(BITRE_REPO+airline_activity_path, sheet_name='Data')

# Get rid of last column
airline_activity = airline_activity.iloc[:,:-1]

# set the index to the date column
airline_activity['Month'] = pd.to_datetime(airline_activity['Month'])
airline_activity['Month'] = airline_activity['Month'].dt.date
airline_activity.set_index('Month', inplace=True)

airline_activity.head()
```

```python
airline_activity.to_csv("../raw-csvs/bitre/airline_activity.csv")
```
