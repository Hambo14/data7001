```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## ABS Data

### ABS Overseas Arrivals

```python
ABS_REPO = "../raw-data/ABS_Overseas_Arrivals/"

arrival_by_citizenship_path = 'arrivals_country_of_citizenship.xlsx'
arrivals_by_citizenship = pd.read_excel(ABS_REPO+arrival_by_citizenship_path, sheet_name='Table 14.1')

arrivals_by_citizenship.iloc[:,12:].head()
```

```python
# Make the 11th row the header
headers = arrivals_by_citizenship.iloc[11]
arrivals_by_citizenship.columns = headers

# Find where the actual data starts and ends
arrivals_by_citizenship = arrivals_by_citizenship.iloc[12:,:]
arrivals_by_citizenship = arrivals_by_citizenship.iloc[:-2]

arrivals_by_citizenship.head()
```

```python
# set the index to the date column
arrivals_by_citizenship['Month'] = pd.to_datetime(arrivals_by_citizenship['Month'])
arrivals_by_citizenship['Month'] = arrivals_by_citizenship['Month'].dt.date
arrivals_by_citizenship.set_index('Month', inplace=True)
arrivals_by_citizenship.head()
```

```python
# Save it to csv
arrivals_by_citizenship.to_csv("../raw-csvs/arrivals_by_citizenship.csv")
```

### Arrivals by Visa Group

```python
# Arrivals visa group
arrival_by_visa_group = 'arrivals_visa_group.xlsx'

states = ['NSW','VIC','QLD','SA','WA','TAS','NT','ACT','AUS']
for i, state in enumerate(states):
        
    df = pd.read_excel(ABS_REPO+arrival_by_visa_group, sheet_name=i+1)
    
    # Make the 11th row the header
    if state == 'AUS':
        headers = df.iloc[15]
        df.columns = headers

        # Find where the actual data starts and ends
        df = df.iloc[16:]
        df = df.iloc[:-2]

    else:
        headers = df.iloc[14]
        df.columns = headers
    
        # Find where the actual data starts and ends
        df = df.iloc[15:]
        df = df.iloc[:-2]

    # set the index to the date column
    df['Month'] = pd.to_datetime(df['Month'])
    df['Month'] = df['Month'].dt.date
    df.set_index('Month', inplace=True)

    # save csv
    df.to_csv(f"../raw-csvs/{state}_arrival_by_visa_group.csv")
    
```

### Total Arrivals

```python
# Total Arrivals
total_arrival_path = 'arrivals.xlsx'
total_arrival = pd.read_excel(ABS_REPO+total_arrival_path, sheet_name=1)

total_arrival.rename(columns={'Unnamed: 0': 'Month'}, inplace=True)

# Find where the actual data starts and ends
total_arrival = total_arrival.iloc[10:,:-1]

total_arrival.columns = ['Month'] + [header.split(';')[1].strip() for header in total_arrival.columns if not header=='Month']

# set the index to the date column
total_arrival['Month'] = pd.to_datetime(total_arrival['Month'])
total_arrival['Month'] = total_arrival['Month'].dt.date
total_arrival.set_index('Month', inplace=True)
total_arrival.head()
```

```python
total_arrival.to_csv('../raw-csvs/total_arrivals.csv')
```
