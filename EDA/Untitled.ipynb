```python
import pandas as pd
```

```python
BITRE_PATH = "../raw-csvs/bitre/"

seat_utilisation_df = pd.read_csv(BITRE_PATH+'seat_utilisation.csv')
airline_activity = pd.read_csv(BITRE_PATH+'airline_activity.csv')
```

```python
seat_utilisation_df.set_index("MONTH", inplace=True)
```

```python
airline_activity.set_index("Month", inplace=True)
```

```python
seat_utilisation_df.rename(columns={"AIRLINE": "Scheduled Operator", "COUNTRY": "Country to/from"}, inplace=True)
seat_utilisation_df.index.names = ['Month']
seat_utilisation_df.head()
```

```python
new_df = pd.merge(seat_utilisation_df,airline_activity, how='left',on=["Month", "Scheduled Operator", "Country to/from"])
```
